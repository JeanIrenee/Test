```python
#this Range Kutta code was done during the skills phase in python c 
def RK4(F,beta,t):
    h = t[1]-t[0]
    x = []
    y = []
    x.append(beta[0])
    y.append(beta[0])
#     X = np.empty((len(t),len(beta)))
#     X[0] = beta
    k = 0
    while k<(len(t)-1):
        i = 0
        k1 = F(x[k],y[k]) #in this case we don't need the time vector hence we can ignore it in that code
        k2 = F(x[k]+(h/2)*k1[i],y[k]+(h/2)*k1[i+1])
        k3 = F(x[k]+(h/2)*k2[i],y[k]+(h/2)*k2[i+1])
        k4 = F(x[k]+(h)*k3[i],y[k]+(h)*k3[i+1])
        
        x.append(x[k]+(h/6)*(k1[i] + 2*k2[i] + 2*k3[i] + k4[i]))
        y.append(y[k]+(h/6)*(k1[i+1] + 2*k2[i+1] + 2*k3[i+1] + k4[i+1]))
        k += 1 
    return x,y
```

# Exercise 1.1

```python
def F1(a,b,t=0):
#     x,y=Z
    return np.array([5*a + b,3*a + b])
```

```python
t = np.linspace(0,5)
fig2 = plt.figure(figsize=(8,6))
for i in np.linspace(-2,2,10):
    for k in np.linspace(-2,2,10):
        beta = [i,k]
        x,y=RK4(F1,beta,t)
        plt.plot(x,y)

plt.title('The phase diagram: Improper node')#two indep. eigenvectors
plt.xlabel('x(t)') #unstable
plt.ylabel('v(t)')

# we impove the xlim and ylim
plt.xlim(-2, 2)
plt.ylim(-2, 2)

#we add the horizontale and verticale lines
plt.hlines(0, -2, 2,'black')
plt.vlines(0, -2, 2,'black')

# Grid and Arrows
A,B = np.meshgrid(np.linspace(-2,2,10),np.linspace(-2,2,10))

a = 5*A + B
b = 3*A + B

#Normalize the arrows
N = np.sqrt(a**2+b**2)  
U, V = a/N, b/N

pl.quiver(A, B, U, V, color = "black")
plt.grid()
plt.show()
```
