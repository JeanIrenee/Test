```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

# General Formula 

```python
def RK4Method(F,beta,t): #beta is initial condition vector
    h = t[1]-t[0]
    X = np.zeros((len(t),len(beta)))
    X[0] = beta
    k = 0
    while k<(len(t)-1):
        k1 = F(X[k],t[k])
        k2 = F(X[k]+(h/2)*k1,t[k]+(h/2))
        k3 = F(X[k]+(h/2)*k2,t[k]+(h/2))
        k4 = F(X[k]+(h)*k3,t[k]+h)
        
        X[k+1] = X[k]+(h/6)*(k1 + 2*k2 + 2*k3 + k4)
        k += 1 
    return X
```

```python
def RK4(F,beta,t):
    h = t[1]-t[0]
    x=[]
    y=[]
    x.append(beta[0])
    y.append(beta[1])
    k = 0
    while k<(len(t)-1):
        k1 = F(x[k],y[k],t[k])
        k2 = F(x[k]+(h/2)*k1[0],y[k]+(h/2)*k1[1],t[k]+(h/2))
        k3 = F(x[k]+(h/2)*k2[0],x[k]+(h/2)*k2[0],t[k]+(h/2))
        k4 = F(x[k]+(h)*k3[0],y[k]+(h)*k3[1],t[k]+h)
        
        x.append(x[k]+(h/6)*(k1[0] + 2*k2[0] + 2*k3[0] + k4[0]))
        y.append(y[k]+(h/6)*(k1[1] + 2*k2[1] + 2*k3[1] + k4[1]))
        k += 1 
    return x,y
```

# Ex 1.1

```python
def F1(x,y,t): 
    F = np.empty(2)
    F[0] = 5*x + y
    F[1] = 3*x + y
    return F
```

```python
t = np.linspace(-4, 4, 100)
fig = plt.figure(1,figsize=(12,8))
for i in np.linspace(-1,1,10):
    for j in np.linspace(-1,1,10):
        beta = [i,j]
        x,y = RK4(F1,beta,t)
        plt.plot(x,y,label=beta)
#     plt.legend(loc="upper right")

plt.title('The phase diagram')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel("x(t)")
plt.ylabel("v(t)")

x,y = np.mgrid[-4:4:10j, -4:4:10j]
u =5*x + 1*y
v = 3*x + 1*y


plt.grid()
plt.quiver(x, y, u, v, color = 'b',pivot='mid')

plt.show()
```
