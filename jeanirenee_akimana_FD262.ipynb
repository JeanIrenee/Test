```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
def RK4(t,F,beta):
    h = t[1]-t[0]
    x=[]
    y=[]
    x.append(beta[0])
    y.append(beta[1])
    k = 0
    while k<(len(t)-1):
        k1 = F(x[k],y[k],t[k])
        k2 = F(x[k]+(h/2)*k1[0],y[k]+(h/2)*k1[1],t[k]+(h/2))
        k3 = F(x[k]+(h/2)*k2[0],x[k]+(h/2)*k2[0],t[k]+(h/2))
        k4 = F(x[k]+(h)*k3[0],y[k]+(h)*k3[1],t[k]+h)
        
        x.append(x[k]+(h/6)*(k1[0] + 2*k2[0] + 2*k3[0] + k4[0]))
        y.append(y[k]+(h/6)*(k1[1] + 2*k2[1] + 2*k3[1] + k4[1]))
        k += 1 
    return x,y
```

```python
def F2(x1,y1,t):
#     x,y = X
    F = np.empty(2)
    F[0] = 2*x1 + -5*y1
    F[1] = x1 + -2*y1
    return F
```

```python
#plot
fig2 = plt.figure(figsize=(8,6))
ax4 = fig2.add_subplot(1,1,1)
for i in range(100):
    beta = np.random.randint(-2, 2, 2).round()
#     beta=np.random.randn(2).round()
    x,y = RK4(t,F2,beta) #odeint(F1,beta,t)#
    ax4.plot(x,y)
#     plt.legend(loc="upper right")
# quiverplot
# define a grid and compute direction at each point
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)

X1 , Y1  = np.meshgrid(x, y)# create a grid
# DX1, DY1 = F1(X1, Y1,t)     # compute growth rate on the grid
DX1 = 2*X1 + -5*Y1
DY1 = X1 - 2*Y1
M = (np.hypot(DX1, DY1))                        # norm growth rate 
# M[ M == 0] = 1.                                 # avoid zero division errors 
# DX1 /= M                                        # normalize each arrows
# DY1 /= M

ax4.quiver(X1, Y1, DX1, DY1)
ax4.legend()
ax4.grid()
```
