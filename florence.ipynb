```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace(0.01,1,100)
T = np.linspace(0.01,100,100)
E = np.linspace(-0.9,1,100)
E1 = np.linspace(0,5,100)
```

```python
def s(p):
    return -p*np.log(p)

def g(T):
    return np.exp(-1/T)

def b(E):
    b_3 = 1/(np.exp(3*(E1+1))-1)
    b_30 = 1/(np.exp(30*(E1+1))-1)
    return b_3,b_30

def f(E):
    f_10 = 1/(np.exp(10*(E1-1))+1)
    f_100 = 1/(np.exp(100*(E1-1))+1)
    return f_10,f_100

b_3,b_30 = b(E)
f_10,f_100 = f(E1)
```

```python
fig= plt.figure(figsize=(10,12))

plt.subplot(3,2,1)
plt.plot(p, s(p))
plt.xlim(0, 1)
plt.ylim(0, 0.4)
plt.xlabel('p')
plt.ylabel('s')
plt.legend(['s'])

plt.subplot(3,2,2)
plt.plot(T, g(T),'r')
plt.xlim(-1, 60)
plt.ylim(0, 1)
plt.xlabel('T')
plt.ylabel('g')
plt.legend(['g'])

plt.subplot(3,2,3)
plt.plot(E,b_3)
plt.plot(E, b_30,'r')
plt.xlabel('E')
plt.ylabel('b_3 - b_30')
plt.legend(['b_3','b_30'])

plt.subplot(3,2,4)
plt.plot(E1,f_10 ,'b')
plt.plot(E1,f_100 ,'r')
plt.xlabel('E')
plt.ylabel('f_10 - f_100')
plt.legend(['f_10','f_100'])


plt.show()
```
