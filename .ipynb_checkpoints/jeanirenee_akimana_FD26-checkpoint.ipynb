```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from pylab import *
```

# General Formula 

```python
def RK4Method(t,F,beta): #beta is initial condition vector
    h = t[1]-t[0]
    X = np.zeros((len(t),2))
    
    X[0] = beta
    k = 0
    while k<(len(t)-1):
        k1 = F(X[k],t[k])
        k2 = F(X[k]+(h/2)*k1,t[k]+(h/2))
        k3 = F(X[k]+(h/2)*k2,t[k]+(h/2))
        k4 = F(X[k]+(h)*k3,t[k]+h)
        
        X[k+1] = X[k]+(h/6)*(k1 + 2*k2 + 2*k3 + k4)
        k += 1 
    return X
```

# Exercise 1

# Ex 1.1

```python
def F1(X,t): 
    F = np.empty(2)
    F[0] = 5*X[0] + X[1]
    F[1] = 3*X[0] + X[1]
    return F
```

```python
t = np.linspace(-4, 4, 100)
fig = plt.figure(1,figsize=(12,8))
for i in range(20):
    beta = np.random.randint(-2, 2, 2).round()
#     beta=np.random.randn(2).round()
    X = RK4Method(t,F1,beta) #odeint(F1,beta,t)#
    plt.plot(X[:,0],X[:,1],label=beta)
    plt.legend(loc="upper right")

plt.title('The phase diagram')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel("x(t)")
plt.ylabel("v(t)")

x,y = np.mgrid[-4:4:10j, -4:4:10j]
u =5*x + 1*y
v = 3*x + 1*y


plt.grid()
plt.quiver(x, y, u, v, color = 'b')

plt.show()
```

# Ex 1.2

```python
def F2(X,t):
    F = np.empty(2)
    F[0] = 2*X[0] + -5*X[1]
    F[1] = 1*X[0] + -2*X[1]
    return F
```

```python
t = np.linspace(-4, 4, 100)
fig = plt.figure(1,figsize=(12,8))
for i in range(20):
    beta = np.random.randint(-2, 2, 2).round()
#     beta=np.random.randn(2).round()
    X = RK4Method(t,F2,beta)
    plt.plot(X[:,0],X[:,1],label=beta)
    plt.legend(loc="upper right")

plt.title('The phase diagram : centre')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel("x(t)")
plt.ylabel("v(t)")

x,y = np.mgrid[-4:4:10j, -4:4:10j]
u =2*x + -5*y
v = 1*x + -2*y


plt.grid()
plt.quiver(x, y, u, v, color = 'b')

plt.show()
```
