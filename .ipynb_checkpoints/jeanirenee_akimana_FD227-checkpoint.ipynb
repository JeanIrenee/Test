```python
#this Range Kutta code was done during the skills phase in python c 
def RK4(F,beta,t):
    h = t[1]-t[0]
    X = np.empty((len(t),len(beta)))
    X[0] = beta
    k = 0
    while k<(len(t)-1):
        k1 = F(X[k]) #in this case we don't need the time vector hence we can ignore it in that code
        k2 = F(X[k]+(h/2)*k1) 
        k3 = F(X[k]+(h/2)*k2) 
        k4 = F(X[k]+(h)*k3)
        
        X[k+1] = X[k]+(h/6)*(k1 + 2*k2 + 2*k3 + k4)
        k += 1 
    return X
```

# Exercise 1.1

```python
def F1(Z,t=0):
    x,y=Z
    return np.array([5*x + y,3*x + y])
```

```python
t = np.linspace(-5,5)
fig2 = plt.figure(figsize=(8,6))
for i in np.linspace(-2,2,10):
    for k in np.linspace(-2,2,10):
        beta = [i,k]
        x=RK4(F1,beta,t)
        plt.plot(x[:,0],x[:,1])

plt.title('The phase diagram: Improper node')#two indep. eigenvectors
plt.xlabel('x(t)') #unstable
plt.ylabel('v(t)')

# we impove the xlim and ylim
plt.xlim(-2, 2)
plt.ylim(-2, 2)

#we add the horizontale and verticale lines
plt.hlines(0, -2, 2,'black')
plt.vlines(0, -2, 2,'black')

# Grid and Arrows
X,Y = np.meshgrid(np.linspace(-2,2,20),np.linspace(-2,2,20))

u = 5*X + Y
v = 3*X + Y

#Normalize the arrows
N = np.sqrt(u**2+v**2)  
U, V = u/N, v/N

pl.quiver(X, Y, U, V, color = "black")
plt.grid()
plt.show()
```

# Exercise 1.2

```python
def F2(Z,t=0):
    x,y=Z
    return np.array([2*x - 5*y,x-2*y])
```

```python
fig2 = plt.figure(figsize=(8,6))
for i in np.linspace(-2,2,10):
    for k in np.linspace(-2,2,3):
        beta = [i,k]
        x=RK4(F2,beta,t)
        plt.plot(x[:,0],x[:,1]) #,label='beta=(%.f, %.f)' % ( beta[0], beta[1]) 

plt.title('The phase diagram: centre') #stable
plt.xlabel('x(t)')
plt.ylabel('v(t)')

# we impove the xlim and ylim
plt.xlim(-2, 2)
plt.ylim(-2, 2)

#we add the horizontale and verticale lines
plt.hlines(0, -2, 2,'black')
plt.vlines(0, -2, 2,'black')


X,Y = np.meshgrid(np.linspace(-2,2,20),np.linspace(-2,2,20))
u = 2*X - 5*Y
v = X - 2*Y

N = np.sqrt(u**2+v**2)  
U, V = u/N, v/N

pl.quiver(X, Y, U, V, color = "black")
plt.grid()
plt.show()
```

# Exercise 1.4

```python
def F4(Z,t=0):
    x,y=Z
    return np.array([2*x - y,3*x-2*y])
```

```python
fig2 = plt.figure(figsize=(8,6))
for i in np.linspace(-2,2,10):
    for k in np.linspace(-2,2,6):
        beta = [i,k]
        x=RK4(F4,beta,t)
        plt.plot(x[:,0],x[:,1])

plt.title('The phase diagram: Saddle point')
plt.xlabel('x(t)')
plt.ylabel('v(t)')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.hlines(0, -2, 2,'black')
plt.vlines(0, -2, 2,'black')

X,Y = np.meshgrid(np.linspace(-2,2,20),np.linspace(-2,2,20))
u = 2*X - Y
v = 3*X - 2*Y

N = np.sqrt(u**2+v**2)  
U, V = u/N, v/N

pl.quiver(X, Y, U, V, color = "black")

plt.grid()
plt.show()
```
