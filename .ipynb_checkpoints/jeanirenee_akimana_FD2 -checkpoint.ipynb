```python
# this Range Kutta code was done during the skills phase in python class. I just try to improve. 
# the idea is to build a Range kutta code which is close to odeint a python built in function
# it takes the same parameter as what we will do in the next step:
def RK4(F,beta,t):                                #F function, beta the initial condition and t the vector time
    h = t[1]-t[0]                                 #stepsize
    X = np.empty((len(t),len(beta)))              # define a a matrix which will contain the solutions
    X[0] = beta                                   #update the first index by the initial condition
    k = 0
    while k<(len(t)-1):                           # One can also use the for loop 
        k1 = np.array(F(X[k],t[k]))               #we add np.array and we let the user to define its function which can return a list
        k2 = np.array(F(X[k]+(h/2)*k1,t[k]+(h/2))) # Because operation with a list doesn't work in python
        k3 = np.array(F(X[k]+(h/2)*k2,t[k]+(h/2))) # we have to convert a list in array.
        k4 = np.array(F(X[k]+(h)*k3,t[k]+h))
        
        X[k+1] = X[k]+(h/6)*(k1 + 2*k2 + 2*k3 + k4) #after getting the $k_i$ coefficients we update X[k+1]
        k += 1 #update k
    return X
```

```python
t = np.linspace(0,5,200) #time vector
```

# Exercise 1.1

```python
def F1(Z,t=0):
    return [5*Z[0] + Z[1],3*Z[0] + Z[1]] #the function returns a list
```

```python
fig2 = plt.figure(figsize=(8,6))

#plot the trajectories in forward in time       
[plt.plot(RK4(F1,[i,k],t)[:,0],RK4(F1,[i,k],t)[:,1],'b-') 
 for i in np.linspace(-2,2,10) for k in np.linspace(-2,2,3)]

#plot the trajectories in backward in time   
[plt.plot(RK4(F1,[i,k],-t)[:,0],RK4(F1,[i,k],-t)[:,1],'b-') 
 for i in np.linspace(-2,2,10) for k in np.linspace(-2,2,3)]


plt.title('The phase diagram: Improper node')#two indep. eigenvectors
plt.xlabel('x(t)') #unstable
plt.ylabel('v(t)')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.hlines(0, -2, 2,'black')
plt.vlines(0, -2, 2,'black')

# We define a grid and we compute the direction of the trajectories 
# Below we define a grid
X,Y = np.meshgrid(np.linspace(-2,2,10),np.linspace(-2,2,10))

a,b = F1([X,Y])            #compute the F1 function on the grid


Norm = np.sqrt(a**2+b**2)  #Norm of the F1 fuction
A = a/Norm                 # Normalize arrows 
B = b/Norm

# To drow the direction of trajectories we use quiver function
pl.quiver(X, Y, A,B, color = "black")
plt.grid()
plt.show()
```
