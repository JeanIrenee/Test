```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace(0.0001,1,100)
T = np.linspace(0.0001,100,100)
E1 = np.linspace(-0.9999,1,100)
E2 = np.linspace(0,5,100)
```

```python
fig= plt.figure(figsize=(10,12))

plt.subplot(3,2,1)
plt.plot(p, -p*np.log(p)) #s(p)
plt.xlim(0, 1)
plt.ylim(0, 0.4)
plt.xlabel('p-axis')
plt.ylabel('s(p)-axis')
plt.legend(['s(p)'])
plt.grid()

plt.subplot(3,2,2)
plt.plot(T, np.exp(-1/T),'r')#g(T)
plt.xlim(-1, 60)
plt.ylim(0, 1)
plt.xlabel('T-axis')
plt.ylabel('g(T)-axis')
plt.legend(['g(T)'])
plt.grid()

plt.subplot(3,2,3)
plt.plot(E1,1/(np.exp(3*(E1+1))-1) )#b3(E1)
plt.plot(E1,1/(np.exp(30*(E1+1))-1) ,'r')#b30(E1)
# plt.xlim(-2, 20)
plt.ylim(0, 15)
plt.xlabel('E-axis')
plt.ylabel('b3 and b30(E)-axis')
plt.legend(['b3','b30'])
plt.grid()

plt.subplot(3,2,4)
plt.plot(E2,1/(np.exp(10*(E2-1))+1) ,'b')#f10(E2)
plt.plot(E2,1/(np.exp(100*(E2-1))+1) ,'r')#f100(E2)
# plt.xlim(-1, 10)
# plt.ylim(0, 1)
plt.xlabel('E-axis')
plt.ylabel('f10 and f100(E)-axis')
plt.legend(['f10','f100'])


plt.grid()
plt.show()
```
