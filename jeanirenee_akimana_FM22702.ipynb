```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
# from pylab import *
import pylab as pl
```

```python
RH = np.arange(0,100+0.01,0.1)
```

```python
def OpportunitySet(RH):
    sigmaH = (1/400)*(210000*(RH**2) - 42000*RH + 2356)
    return sigmaH
```

```python
fig = plt.figure(1,figsize=(12,8))
plt.plot(RH,OpportunitySet(RH))
plt.title(' Risk vs Expected returns')

plt.xlim(0, 50,10)
# plt.ylim(-4, 4)
plt.xlabel(" Expected returns")
plt.ylabel("Risk")



plt.grid()

plt.show()
```
