# 1. Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def EulerMethod(a,b,c,N,f):
#     h = (b-a)/N
    x = np.linspace(a,b,N+1)
    h = x[1] - x[0]
    y = np.zeros(N+1) 
#     y = np.zeros(len(x)) 
    y[0] = c
    for i in range(0,N):
        y[i+1] = y[i] + h*f(y[i],x[i])
    return y
```

```python
def EulerMethod1(f,c,t):

    h = t[1] - t[0]
    y = np.zeros(len(t)) 
#     y = np.zeros(len(x)) 
    y[0] = c
    for i in range(0,len(t)-1):
        y[i+1] = y[i] + h*f(y[i],t[i])
    return y
```

```python
a = 0
b = 1
c = 1/9
N = 100

def f(y,x):
    return np.exp(x)/(3+6*np.exp(x))

# g = lambda y,x: np.exp(x)/(3+6*np.exp(x))


EulerMethod(a,b,c,N,f)
```

```python
t = np.linspace(a,b,N+1)
EulerMethod1(f,c,t)
```

```python
def RangeKutta2(a,b,c,N,f):
#     h = (b-a)/N
    x = np.linspace(a,b,N+1)
    h = x[1] - x[0]
    y = np.zeros(N+1) 
#     y = np.zeros(len(x)) 
    y[0] = c
    for i in range(0,N):
        k1 = h*f(y[i],x[i])
        k2 = h*f(y[i]+k1,x[i]+h)
        y[i+1] = y[i] + (1/2)*(k1+k2)
    return y
```

```python
RangeKutta2(a,b,c,N,f)
```

```python
def RangeKutta4(a,b,c,N,f):
#     h = (b-a)/N
    x = np.linspace(a,b,N+1)
    h = x[1] - x[0]
    y = np.zeros(N+1) 
#     y = np.zeros(len(x)) 
    y[0] = c
    for i in range(0,N):
        k1 = h*f(y[i],x[i])
        k2 = h*f(y[i]+(k1/2),x[i]+(h/2))
        k3 = h*f(y[i]+(k2/2),x[i]+(h/2))
        k4 = h*f(y[i]+k3,x[i]+h)
        y[i+1] = y[i] + (1/6)*(k1+2*k2+2*k3+k4)
    return y
```

```python
RangeKutta4(a,b,c,N,f)
```

```python
def EulerMethodSyst(a,b,c,N,f):
#     h = (b-a)/N
    t = np.linspace(a,b,N+1)
    h = t[1] - t[0]
    z = np.zeros((N+1,len(c))) 
#     y = np.zeros(len(x)) 
    z[0,] = c
    for i in range(0,N):
        z[i+1,] = z[i,] + h*f(z[i,],t[i])
    return z
```
